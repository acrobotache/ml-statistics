# Measure of Central Tendancy
The mean, median and mode are all valid measures of central tendency, but under different conditions, some measures of central tendency become more appropriate to use than others. 

# MEAN
The arithmatic mean also known as arithmatic average is a central value of finite set of numbers in a dataset,  specifically, the sum of the values divided by the number of values.

Mean Calculation:

![title](misc/mean_img.svg)

### Merits
1. Fast and Easy to calculate
2. It cannot be determined graphycally. 
2. The data is not required to be arranged in ascending or descending order as it is done when computing the median.
3. It is affected by the value of every item in the series. Therefore the mean can be said to truly represent the central tendency of the data.

### Demerits
1. Sensitive to outliers (extream low or high values), It may give a a distorted picture of the data due to the presence of extreme values.
2. In an extremely asymmetrical/skewed distribution the mean is usually not a suitable measure for the “central value” of the data.
3. It cannot be located graphically.

# MEDIAN
The median is the value separating the higher half from the lower half of a data sample, a population, or a probability distribution. (SORT AND find middle value)

### Merits
1. It is not affected by extreme values and also interdependent of range or dispersion of the data.
2. It can be determined graphically.
3. Useful in case of skewed data. 

### Demerits
1. Data needs to be sorted first. 
2. It is a less representative average because it does not depend on all the items in the series. 
3. It is affected more by sampling fluctuations than the mean as it is concerned with on1y one item i.e. the middle item. 

# MODE
The mode is the value that appears most often in a set of data values. If X is a discrete random variable, the mode is the value x (i.e, X = x) at which the probability mass function takes its maximum value. In other words, it is the value that is most likely to be sampled. 

Steps to calculate mode:
    1. Collect and organize the data from a dataset.
    2. Determine all the distinct values in a dataset.
    3. Count the frequency of occurrence for each distinct value.
    4. The most frequent value(s) is the mode.
    
### Merits
1. mode is not affected by extremes of values
2. When data are not complete, mode cannot be difficult to estimate
3. It can be located graphically.

### Demerits
1. It is difficult to calculate, especially when more than one mode or large numbers are involved. Mode is ill-defined. It is not always possible to find a clearly defined mode. In some cases we may come across distribution are called bi-modal. If a distribution has more than two modes, it is said to be multimodal. 
2. It is not based on all values.
3. As compared with mean, mode is affected to a greater extent, by fluctuations of sampling.

## Hackerrank Problem Statement on Mean, Median and Mode
[Problem Statement](https://www.hackerrank.com/challenges/s10-basic-statistics/problem?isFullScreen=true)

Solution:

```python
N = int(input())
X = list(map(int, input().split()))

# calulate mean
mean = sum(X)/N

# calculate median 
X.sort()
if not N%2:
    median = (X[N//2 -1] + X[N//2])/2.0
else:
    median = X[N//2]

# calculate mode
occ_dict = dict()
for i in X:
    if i not in occ_dict.keys():
        occ_dict[i]=1
    else:
        occ_dict[i] += 1
mode=0
max_occ = 0
for key, value in occ_dict.items():
    if value > max_occ:
      mode = key
      max_occ = value  
      
# print mean, median and mode
print(mean)
print(median)
print(mode)
```
